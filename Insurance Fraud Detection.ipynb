```python
# Handling missing values
import pandas as pd

# Replacing '?' with NaN for easy handling
df.replace('?', pd.NA, inplace=True)

# Handling missing values in 'collision_type', 'property_damage', and 'police_report_available'
# Replacing with mode for categorical columns and set default for 'property_damage'
df['collision_type'].fillna(df['collision_type'].mode()[0], inplace=True)
df['property_damage'].fillna('Unknown', inplace=True)
df['police_report_available'].fillna(df['police_report_available'].mode()[0], inplace=True)

# Converting categorical columns 
df['collision_type'] = df['collision_type'].astype('category')
df['property_damage'] = df['property_damage'].astype('category')
df['police_report_available'] = df['police_report_available'].astype('category')
```

### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df.drop(columns=['policy_bind_date'], inplace=True)
```

```python
# List of object columns to convert
object_columns = df.select_dtypes(include=['object']).columns

# Initializing LabelEncoder
label_encoders = {}

for col in object_columns:
    label_encoders[col] = LabelEncoder()
    df[col] = label_encoders[col].fit_transform(df[col].astype(str))
```

### Standard Scaler

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

target = df['fraud_reported']
features = df.drop(columns=['fraud_reported'])

numerical_columns = features.select_dtypes(include=['int64', 'float64']).columns
scaler = StandardScaler()
features[numerical_columns] = scaler.fit_transform(features[numerical_columns])
df_scaled = pd.concat([features, target], axis=1)
print(df_scaled.head())
```

### Treating Class Imabalnce

```python
# Defining a function to convert specified categorical columns to numeric using LabelEncoder
def convert_categorical_to_numeric(df_features, df_target):
    # Initialize LabelEncoder
    le = LabelEncoder()
    
    # Converting categorical columns in features dataframe to numeric
    categorical_cols = ['collision_type', 'property_damage', 'police_report_available']
    for col in categorical_cols:
        df_features[col] = le.fit_transform(df_features[col])
    
    # Converting target column to numeric if needed (only if it's not already numeric)
    if df_target.dtype == 'object':
        df_target = le.fit_transform(df_target)
    
    return df_features, df_target

features, target = convert_categorical_to_numeric(features, target)
print("Features data types:")
print(features.dtypes)
print("\nTarget data type:")
print(target.dtype)
```

```python
from imblearn.over_sampling import SMOTE

# Applying SMOTE to handle class imbalance
smote = SMOTE(random_state=42)
features_resampled, target_resampled = smote.fit_resample(features, target)

# Combining features and target back into a single dataframe
df_resampled = pd.concat([features_resampled, target_resampled], axis=1)

# Display the balanced dataframe
print(df_resampled['fraud_reported'].value_counts())
```

### Model Training and Evaluation

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Initializing and training the Random Forest model
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# Predicting on the test data and evaluating the model
y_pred = rf_model.predict(X_test)

# Calculating accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Printing classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Printing confusion matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

### Saving the model

```python
import joblib

model_filename = 'random_forest_model.pkl'
joblib.dump(rf_model, model_filename)

print(f"Trained Random Forest model saved as {model_filename}")
```
